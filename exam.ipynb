```python
df['total_score']=(df['math score']+df['reading score']+df['writing score'])
```

```python
df['avg_score']=df['total_score']/3
```

```python
df['avg_score'].min()
```

```python
fig,axis=plt.subplots(1,2,figsize=(15,17))
plt.subplot(121)
sns.histplot(data=df,x='avg_score',bins=30,kde=True)
plt.subplot(122)
sns.histplot(data=df,x='total_score',bins=30,kde=True,hue='gender')
```

```python
plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='avg_score',kde=True,hue='lunch')
plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],x='avg_score',kde=True,hue='lunch')
plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],x='avg_score',kde=True,hue='lunch')
```
